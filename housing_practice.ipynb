```python
import pandas as pd
housing_data = pd.read_csv('Housing.csv')
housing_data.head()
```

```python
housing_data.isnull().sum()
```

```python
housing_data["date"]= pd.to_datetime(housing_data['date'])
housing_data.head()
```

```python
conda update --all
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

resampled_data = housing_data.resample('M', on='date').mean().reset_index()

plt.figure(figsize=(15,8))
sns.lineplot(x='date', y='price', data=resampled_data, color='red')
plt.show()
```
