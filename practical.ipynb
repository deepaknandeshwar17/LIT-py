# ***Gradient Descent***
---

```python
import numpy as np
import matplotlib.pyplot as plt

# Data points (x, y)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 2.8, 3.6, 4.5, 5.1])

# Hypothesis function
def hypothesis(theta_0, theta_1, x):
    return theta_0 + theta_1 * x

# Cost function
def cost_function(theta_0, theta_1, x, y):
    m = len(y)
    return (1 / (2 * m)) * np.sum((hypothesis(theta_0, theta_1, x) - y) ** 2)

# Gradient Descent with step tracking
def gradient_descent_with_steps(x, y, theta_0, theta_1, alpha, iterations):
    m = len(y)
    history = []  # To store values after each iteration

    for i in range(iterations):
        # Compute gradients
        gradient_theta_0 = (1 / m) * np.sum(hypothesis(theta_0, theta_1, x) - y)
        gradient_theta_1 = (1 / m) * np.sum((hypothesis(theta_0, theta_1, x) - y) * x)

        # Update parameters
        theta_0 -= alpha * gradient_theta_0
        theta_1 -= alpha * gradient_theta_1

        # Record the current state of parameters and cost
        current_cost = cost_function(theta_0, theta_1, x, y)
        history.append((i + 1, theta_0, theta_1, current_cost))

    return theta_0, theta_1, history

# Parameters for Gradient Descent
theta_0_initial = 0
theta_1_initial = 0
alpha = 0.01  # Learning rate
iterations = 20  # Number of iterations for better visualization

# Perform gradient descent
theta_0_final, theta_1_final, steps_history = gradient_descent_with_steps(
    x, y, theta_0_initial, theta_1_initial, alpha, iterations
)

# Extract iterations and costs for visualization
iterations_list = [step[0] for step in steps_history]
cost_list = [step[3] for step in steps_history]


print("Iteration | theta_0      | theta_1      | Cost")
print("-----------------------------------------------")
for step in steps_history:
    print(f"{step[0]:9} | {step[1]:.6f} | {step[2]:.6f} | {step[3]:.6f}")


# Plot the cost function over iterations
plt.figure(figsize=(10, 6))
plt.plot(iterations_list, cost_list, marker='o', label="Cost Function")
plt.title("Cost Function Convergence Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Cost (J)")
plt.grid(True)
plt.legend()
plt.show()

# Print final parameters
print(f"Final parameters: theta_0 = {theta_0_final:.6f}, theta_1 = {theta_1_final:.6f}")
```
